```python
def extraer_net_reserve(path_excel, tolerance=1):
    
    wb = load_workbook(path_excel, data_only=True)
    resultados = []

    for sheet_name in wb.sheetnames:
        ws = wb[sheet_name]

        policy = None
        carrier_value = None
        net_reserve_value = None

        max_row = ws.max_row
        max_col = ws.max_column

        # -------------------------------------------------
        # 🔎 Buscar POLICY N° (LOCAL) o POLICY N°
        # -------------------------------------------------
        targets = ["POLICY N° (LOCAL)", "POLICY N°"]

        for row in ws.iter_rows():
            for cell in row:
                cell_value = (
                    str(cell.value)
                    .replace("\xa0", " ")
                    .strip()
                    .upper()
                    if cell.value else ""
                )

                if cell_value in [t.upper() for t in targets]:
                    if cell.column + 2 <= max_col:
                        policy = ws.cell(row=cell.row, column=cell.column + 2).value
                    break
            if policy is not None:
                break

        # -------------------------------------------------
        # 🔎 Buscar LOCAL CARRIER
        # -------------------------------------------------
        for row in ws.iter_rows():
            for cell in row:
                cell_value = (
                    str(cell.value)
                    .replace("\xa0", " ")
                    .strip()
                    .upper()
                    if cell.value else ""
                )

                if cell_value == "LOCAL CARRIER":
                    if cell.column + 2 <= max_col:
                        carrier_value = ws.cell(row=cell.row, column=cell.column + 2).value
                    break
            if carrier_value is not None:
                break

        # -------------------------------------------------
        # 🔎 Buscar fila donde esté NET RESERVE
        # -------------------------------------------------
        header_row_index = None

        for row in ws.iter_rows(min_row=1, max_row=25):
            for cell in row:
                if cell.value:
                    clean_value = (
                        str(cell.value)
                        .replace("\xa0", " ")
                        .strip()
                        .upper()
                    )

                    if "NET RESERVE (USD)" in clean_value:
                        header_row_index = cell.row
                        break
            if header_row_index:
                break

        # -------------------------------------------------
        # 🔎 Buscar segunda columna NET RESERVE
        # -------------------------------------------------
        if header_row_index is not None:

            headers = []

            for col in range(1, max_col + 1):
                val = ws.cell(row=header_row_index, column=col).value

                if val:
                    clean_val = (
                        str(val)
                        .replace("\xa0", " ")
                        .strip()
                        .upper()
                    )

                    if "NET RESERVE (USD)" in clean_val:
                        headers.append(col)

            if len(headers) >= 2:

                target_col = headers[1]
                numeric_values = []

                for row in range(header_row_index + 1, max_row + 1):
                    val = ws.cell(row=row, column=target_col).value

                    if val is not None:
                        try:
                            num = float(str(val).replace(",", "").replace("$", ""))

                            # 🔥 aplicar tolerancia aquí
                            if abs(num) < tolerance:
                                num = 0.0

                            numeric_values.append(num)

                        except:
                            pass

                if numeric_values:
                    net_reserve_value = numeric_values[-1]

        # -------------------------------------------------
        # Guardar resultados
        # -------------------------------------------------
        resultados.append({
            "Policy Number": policy,
            "Cedant": carrier_value,
            "Cover": sheet_name,
            "Net Reserve": net_reserve_value
        })

    return pd.DataFrame(resultados)
```

```python
df_list = [globals()[f'path_bdx_{i}'] for i in range(6, 14) if f'path_bdx_{i}' in globals()]


df_total = pd.concat([extraer_net_reserve(path) for path in df_list],
                     ignore_index=True)

#RENOMBRAR COVER
df_total["Cover"] = df_total["Cover"].apply(
    lambda x: "TRANSPORTE"
    if isinstance(x, str) and "TRANSPORTE" in x.upper()
    else x
)

df_total["Cover"] = df_total["Cover"].apply(
    lambda x: "FERREO"
    if isinstance(x, str) and "FERREO" in x.upper()
    else x
)

df_total["Cover"] = df_total["Cover"].apply(
    lambda x: "FLETADORES RC"
    if isinstance(x, str) and "FLETADORES RC" in x.upper()
    else x
)

df_total["Cover"] = df_total["Cover"].apply(
    lambda x: "CASCO"
    if isinstance(x, str) and "CASCO" in x.upper()
    else x
)

df_total["Cover"] = df_total["Cover"].apply(
    lambda x: "CARGA"
    if isinstance(x, str) and "CARGA" in x.upper()
    else x
)

df_total["Cover"] = df_total["Cover"].apply(
    lambda x: "PANDI"
    if isinstance(x, str) and "PANDI" in x.upper()
    else x
)

df_total["Net Reserve"]=df_total["Net Reserve"].fillna(0)


# Final database after merge
ruta= rf"{directorio_proyecto2}/Contabilidad/{AñoMes}" 
if not os.path.exists(f'{ruta}'):
    os.makedirs(f'{ruta}')

df_total.to_excel(f'{ruta}/OSLR {AñoMes}.xlsx', index=False)
```
